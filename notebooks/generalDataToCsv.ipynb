```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
```

### **For NBI light source**

```python
path = '../data/NBI_augmented/'
path_adenoma = path + 'adenoma_NBI/'
path_hiperplastic = path + 'hiperplastic_NBI/'
path_serrated = path + 'serrated_NBI/'
videos_adenoma = os.listdir(path_adenoma)
videos_hiperplastic = os.listdir(path_hiperplastic)
videos_serrated = os.listdir(path_serrated)

print("==== videos adenoma ====")
print(videos_adenoma)
print("==== videos hiperplastic ====")
print(videos_hiperplastic)
print("==== videos serrated ====")
print(videos_serrated)
```

```python
csvfile = open('datasetNbiAugmented.csv', '+w')
for video in videos_adenoma:
    current_path = path_adenoma+video+"/"
    files = os.listdir(current_path)
    for file in files:
        csvfile.write(current_path+file+',adenoma\n')

for video in videos_hiperplastic:
    current_path = path_hiperplastic+video+"/"
    files = os.listdir(current_path)
    for file in files:
        print(current_path+file)
        csvfile.write(current_path+file+',hiperplastic\n')
        
for video in videos_serrated:
    current_path = path_serrated+video+"/"
    files = os.listdir(current_path)
    for file in files:
        print(current_path+file)
        csvfile.write(current_path+file+',serrated\n')
        
csvfile.close()
```
