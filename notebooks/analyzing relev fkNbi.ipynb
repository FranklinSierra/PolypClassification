```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import imageio
import shutil
import heapq
import os
```

## Relevant frames

```python
#cambiar el path segun la clase de interes
fk_ade = pd.read_csv('../data/NBI_artificial/predictions/predict-adenomaFakeNbi.csv', header = None)
fk_ade.columns = ['path', 'frame', 'ade_prob', 'hyp_prob', 'ser_prob']
fk_ade.head()
```

```python
clase = 'adenoma_WL'
new_clase = 'adenoma_NBI'
gen_path = '../../cycleGan/data/wl2nbi/fake_images_ck4/'
gen_path2 = '../data/NBI_augmented/' + str(new_clase) + '/video_'
can = 40
for i in range(1, can+1):    
    
    if i < 10:
        print("=== video ", i, " ===")
        current_df = fk_ade[fk_ade['path'].str.contains(clase+'/video_0'+str(i))]        
    else:
        print("=== video ", i, " ===")
        current_df = fk_ade[fk_ade['path'].str.contains(clase+'/video_'+str(i))]
        
    to_add = int(len(current_df)/2)
    print("to add: ", to_add)
    max_ade_prob = np.max(current_df['ade_prob'], axis=0)
    min_ade_prob = np.min(current_df['ade_prob'], axis=0)
    threshold = float((max_ade_prob-min_ade_prob)/2)
    frames = current_df[current_df['ade_prob']>threshold]['frame'].tolist()
    #total de imgs en nbi real
    real_nbi = gen_path2 + str(i)
    total = len(os.listdir(real_nbi))
    j = 0
    cont_add = 0    
    for k in range(len(frames)):
        frame = frames[k]
        cont_add += 1
        if cont_add <= to_add:
            j += 1            
            idx = total + j
            or_name = clase + '_video_' + str(i) + '_img_' + str(frame) + '.png'
            new_name = new_clase + '_video_' + str(i) + '_img_' + str(idx) + '.png'
            path1 = gen_path + clase + '/video_' + str(i) + '/' + or_name 
            path2 = gen_path2 + str(i) + '/' + new_name
            shutil.copy2(path1, path2)
        else:
            k = len(frames) + 1               
```
